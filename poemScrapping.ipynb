```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
def get_poems(url):
    html_text = requests.get(url)
    if html_text.status_code == 200:  # If the request was successful
        soup = BeautifulSoup(html_text.content, 'lxml')  # we red the content of the webpage
        poem_cards = soup.find_all("tr") # we filter out all the entries/cells which have information about poem
        for poem in poem_cards:
            # Check if td exists before accessing html_text
            if poem.td:
                # extract the text from the td element
                # and we extract related info about poet and year and print them
                poem_name = poem.td.get_text(strip=True)
                poet_name = poem.find_all("td")[1].get_text(strip=True)
                poem_year = poem.find_all("td")[2].get_text(strip=True)
                poem_url = poem.find_all("td")[0].find("a")["href"]
                poet_url = poem.find_all("td")[1].find("a")["href"]
                print(f"Poem Name: {poem_name}\n, Poet Name: {poet_name}\n, Poem Year: {poem_year}\n")
                print(f"About poem: https://poets.org{poem_url}")
                print(f"About poet: https://poets.org{poet_url}")
                print("")
            else:
                pass #in itreating over the cells we will hit the first cell which contains the headings
                     #so we pass over the heading to not face an error


url = "https://poets.org/poems?page=0"
# get_poems(url)

# now to itreate over multiple pages we will just run the above function multiple times for different pages
# in order to do that we modify the url and update in every itreation to change the page number
a = int(input("Enter number of pages you want to scrape,page numbers from range from 1 to 783"))
if(a<1 or a>783):
    print("Invalid range")
else:
  print("Scraping...")
  for i in range(a):
      url = f"https://poets.org/poems?page={i}"
      get_poems(url)

```
